## 정리해보자면 판다스에서 row, column indexing의 방법은 다섯가지가 있다
1. key방식  
가장기본적인 방식이다
2. attribute 방식  
편하게 쓰기위해서 판다스에서 지원했지만 사이즈나 공백이 똑같은 이름이 있을때 잘동작하지않음 추천하지않는다
3. fancy indexing  
데이터 프레임으로가져온다 (good)
4. loc 방식  
이름을 맞춰서 가져오는 방식
5. iloc 방식  
인덱스를 맞춰서 가져오는 방식, 인덱스이기 때문에 마지막 꺼 안가져옴 예르들어 3:6 이면 3,4,5 행을 가져온다.  
속도자체는 numpy에서 가져온 key방식이 가장 빠르다고한다

```python
tips = sns.load_dataset('tips')
```

```python
#numpy의 fancy indexing 기능 지원
tips[['day','time']]
```

```python
tips.day
#왼쪽 : index
#오른쪽 : numpy의 array
```

```python
type(tips.day)
```

```python
tips.day.values
```

```python
#numpy의 브로드캐스팅 지원
tips.tip+5
```

```python
#numpy의 reduction 지원
tips.tip.sum() 
```

```python
#행가져오는법
#iloc : 인덱스로가져옴. 끝에있는거 포함X
#loc : 이름으로 가져옴. 끝에애ㅣㅆ는거 포함O
```

```python
tips.iloc[3] #indexlocation, 즉 번호대로 가져옴
```

```python
tips.iloc[3:6]
#numpy기반이므로 슬라이싱 가능
```

```python
tips.loc[3:6]
#얘는 6까지 포함핻줌
```

```python
tips.loc[3,'day'] #3번째의 day rkwudha
```

```python
tips.loc[3:6,['day', 'smoker']]
#콤마도 지원~
```

```python
tips.loc[3:6,['smoker','day']]
#순서바꿔서 가져오는것도 가능
```

```python
tips.iloc[3:6,:-2] #콤마 인덱싱, 콤마슬라이싱 지원
```

```python
tips.iloc[3:6,::-1]
```

```python
#어트리븉방식, 키방식 펜시인덱싱 방식
tips[['tip']]
```

```python
%time tips.tip
```

```python
%time tips[['tip']]
```

```python
%time tips['tip']
#key 방식이 제일 빠르고 문제가 없어.
#쓰기 쫌 귀찮을 뿐
```

```python
%time tips.loc[3, 'tip']
```

```python
%time tips.iat[3,5]
#인덱스로 하나 가져오는애
#항상 i 붙은애가 더 빨라
#상황에따라서 쓴다 ㅋㅋ
```

```python
%time tips.at[3, 'time']
#이름으로 하나 가져오는애
```

```python
#행뽑는애 3개
#열뽑는애 2개  : iloc, loc
#특정값뽑는애 4개 : iloc, loc, iat, at
```
