```python
#지프의 법칙에 부합하는지 확인하는 코드
#상위 10프로, 하위10프로 빼고 나머지가중요하다
#데이터가 편향됐는지 검사하는 코드

import matplotlib.pyplot as plt
from math import log

token = [_[0] for _ in sorted(text.vocab().items(),
                              key=lambda x:x[1], reverse=True)]
freq = [_[1]/text.vocab().most_common(1)[0][1]
       for _ in sorted(text.vocab().items(),
                      key = lambda x:x[1], reverse=True)]
_freq = [log(_) for _ in freq]
rank = [1/(_+1) for _ in range(len(token))]
_rank = [log(_) for _ in rank]
plt.plot([i for i in range(len(token))], freq, "r-")
plt.plot([i for i in range(len(token))], rank, "b-")
plt.show()
plt.plot([i for i in range(len(token))], _freq, "r-")
plt.plot([i for i in range(len(token))], _rank, "b-")
plt.show()
```

**n-gram 모델**  
빈도와 통계 정보를 통해 언어를 해석하는 모델   
단어들의 시퀀스를 통해 확률값을 계산하기 위해 만들어진 랭귀지 모델중 하나이다.  
확률을 통해 다음글자가 무엇인지 예측할 수 있다.  
시퀀스로 몇개를 볼 지?   
n개만큼 보겠다 = n-gram모델  
데이터를 많이 수집하는게 중요하다.    
corpus를 잘 수집해야한다.  
corpus와 데이터는 다르다.  
코퍼스는 어느 한 쪽에 바이어스되지 않고 밸런스가 잘 맞아야 하고, 양이 겁나 많아야함.   
여기서는 그냥 수집한 데이터를 쓸거다.  

```python
sentence = "Life is too short, You need Python"

def eojeol(sentence, N=2):
    tokens = word_tokenize(sentence)  #토큰이 생성되고, 이 토큰을 갖지고 숫자세면됨
    ngram = list()
    for i in range(len(tokens)-(N-1)):
        #i를하나하나씩 증가시키면서 N개만큼 묵ㄲ는다
        #ngram.append(" ".join(tokens[i:i+N]))
        ngram.append(tuple(tokens[i:i+N]))
    return ngram

#음절끼리의 쌍을 만들어줌
def umjeol(token, N=2):
    tokens = list(token)
    ngram = list()
    for i in range(len(tokens)-(N-1)):
        #i를하나하나씩 증가시키면서 N개만큼 묵ㄲ는다
        ngram.append("".join(tokens[i:i+N]))
        #ngram.append(tuple(tokens[i:i+N]))
    return ngram

eojeol(sentence, 3)
umjeol(corpus[0].splitlines()[0])
#short가 나올지 안나올지 튜플들 확률로 분석해서 예측
#어간을 분리할 줒ㄹㅈ 모를때(스템을 어케써야할지 모를때)어간 대신 쓸 수 있다. 
#대신 단점은 엄청나게 많은 쌍을 만들어냄

```

```python
def findBigram(tokens):
    ngram = defaultdict(int)
    for token, freq in tokens.items():
        syllables = token.split()
        for i in range(len(syllables)-1):
            ngram[tuple(syllables[i:i+2])] += freq
    return ngram
```

```python
def mergeMaxkey(ngram, tokens, K):
    maxKey = max(ngram, key=ngram.get)
    maxValue = max(ngram.values())
    newTokens = dict()
    
    for token, freq in tokens.items():
        newTokens[re.sub(" ".join(maxKey),
                         "".join(maxKey), token)
                  if maxValue > K else token] = freq
      
    return newTokens
```
