```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips') #데이터불러옴
```

```python
tips.info() #불러오고 나서 무조건 info
#category : 판다스에서 지원하는 타입
#--> 숫자도 되고 문자도됨. 종류가 없는애들을 카테고리로 지정해놓으면 분석할때 편함
```

```python
tips.describe()
#describe는 숫자데이터만 표현해줌
```

```python
tips
```

```python
tips.boxplot()
#size는 몸집크기. 필요없음
```

```python
tips[['total_bill','tip']].boxplot()
```

```python
#pandas는 aggregation 지원해줌
tips.groupby('sex').mean()
```

```python

tips.groupby('sex').sum()
```

```python

tips.groupby('sex').max()
```

```python

tips.groupby('day').max()
```

```python
tips.groupby(['sex','day','time'])['total_bill','tip'].mean()
#어떤상황에 팁이얼마인지 분석가능
```

```python
t = tips.groupby(['sex'])['total_bill', 'tip'].mean()
```

```python
t.plot.bar()
```
