```python
def download(method,url, param=None, data=None , timeout= 1, maxretries = 3):

    try:
        resp = requests.request(method, url,
                               params= param, data= data,
                                headers={"user-agent":"Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.103 Whale/1.5.73.16 Safari/537.36"})
        resp.raise_for_status()
    except requests.exceptions.HTTPError as e:
#         print(e.code)
#         print(e.reason)
        if 500<= e.response.status_code <600 and maxretries >0:
            print(maxretries)
            time.sleep(timeout)
            download(method,url,param,data,timeout,maxretries-1)
        else:
            print(e.response.status_code)
            print(e.response.reason)
    return resp
```
